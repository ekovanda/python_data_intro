# Dein eigenes Analyseprojekt
Jetzt bist du dran! Suche dir ein einen Datensatz der dich interessiert und ggf. eine oder zwei weitere Personen, mit denen du an diesem Datensatz arbeiten möchtest und wende deine neu erlernten Fähigkeiten darauf an.

Deiner Kreativität und Neugier sind hier keine Grenzen gesetzt. Versuche selbst Fragen zu entwickeln und diese mit Analysen und / oder Visualisierungen bestmöglich zu beantworten.

Als Startpunkt kannst du die Datensätze aus diesem Kurs nehmen:
- `IMDB Top 1000` wenn es um Filme gehen soll
- `Books` wenn du Bücher genauer analysieren möchtest
- `Olympics` wenn Sport dein Ding ist
- `Spotify 2024` falls du einen neuen Datensatz im Kontext von Musik untersuchen willst
- `Board Game Geek` wenn du eine besondere Herausforderung suchst und Brettspiele dein Ding sind.

Alternativ kannst du dich auch selbst nach einem passenden Datensatz umsehen. Beispiele findest du hier:
- [Kaggle Datasets](https://www.kaggle.com/datasets)
- [yfinance](https://www.geeksforgeeks.org/how-to-use-yfinance-api-with-python/)
- [Github Awesome Data](https://github.com/awesomedata/awesome-public-datasets)
- [Statistisches Bundesamt](https://www-genesis.destatis.de/datenbank/online)
- [Data World](https://data.world/search) (erfordert einen Account)
